## Task-2: PREDICTIVE ANALYSIS USING MACHINE LEARNING

# 2.1- BUILD A MACHINE LEARNING MODEL (E.G.,REGRESSION OR CLASSIFICATION) TO PREDICT OUTCOMES BASED ON A DATASET.

```python
# Since there are no missing values, proceed with encoding categorical features and scaling numerical features.

categorical_features = ['state']
df_encoded = pd.get_dummies(df,columns=categorical_features, drop_first=False)

numerical_features = ['colonies_number','yield_per_colony','production','stocks','average_price','year']

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_encoded[numerical_features] = scaler.fit_transform(df_encoded[numerical_features])

df_processed = df_encoded.copy()
display(df_processed.head())
```

```python
# Split the preprocessed data into training and testing sets and print their shapes.

from sklearn.model_selection import train_test_split

x = df_processed.drop('value_of_production', axis=1)
y = df_processed['value_of_production']

X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)

print("Shape of y_test:", y_test.shape)
```

```python
#  Import the LinearRegression model and train it on the training data.

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
# The model has been trained, now I need to evaluate its performance on the test set.

from sklearn.metrics import mean_squared_error, r2_score

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)

print("Mean Squred Error:",mse)
print("R-Square:",r2)
```

```python
# Train a Ridge Regression model as an alternative.

from sklearn.linear_model import Ridge

ridge_model = Ridge()
ridge_model.fit(X_train,y_train)

y_pred_ridge = ridge_model.predict(X_test)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)

print("Ridge Regression Mean Squared Error:", mse_ridge)
print("Ridge Regression R-squared:", r2_ridge)
```

```python
# Train a Lasso Regression model as another alternative.

from sklearn.linear_model import Lasso

lasso_model = Lasso()
lasso_model.fit(X_train, y_train)

y_pred_lasso = lasso_model.predict(X_test)
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)

print("Lasso Regression Mean Squared Error:", mse_lasso)
print("Lasso Regression R-squared:", r2_lasso)
```

```python
# Evaluate the trained model's performance on the testing data using relevant metrics.

from sklearn.metrics import mean_squared_error, r2_score

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

## 2.2- Create a notebook that demonstrates feature selection, model training, and evaluation for a regression task using the dataset

# Feature Selection

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE

model = RandomForestRegressor(n_estimators=100, random_state=42)
rfe = RFE(estimator=model, n_features_to_select=10)
rfe.fit(X_train, y_train)

selected_features_mask = rfe.support_
feature_ranking = rfe.ranking_
selected_features = X_train.columns[selected_features_mask]

print("Selected Features:")
print(selected_features)
```

# Model Tranning


```python
# Create new training and testing sets with selected features, import RandomForestRegressor, instantiate the model, and train it.

X_train_selected = X_train[selected_features]
X_test_selected = X_test[selected_features]

from sklearn.ensemble import RandomForestRegressor

final_model = RandomForestRegressor(n_estimators=100, random_state=42)
final_model.fit(X_train_selected, y_train)
```

# Model evaluation


```python
#  Evaluate the trained model's performance on the testing data using relevant metrics.

from sklearn.metrics import mean_squared_error, r2_score

y_pred_final = final_model.predict(X_test_selected)
mse_final = mean_squared_error(y_test, y_pred_final)
r2_final = r2_score(y_test, y_pred_final)

print("Final Model Mean Squared Error:", mse_final)
print("Final Model R-squared:", r2_final)
```
